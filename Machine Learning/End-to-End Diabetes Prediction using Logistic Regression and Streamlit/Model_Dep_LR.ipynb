```python
import pandas as pd
import numpy as np
import streamlit as st
import warnings
warnings.filterwarnings('ignore')
import pickle
```

```python
model=pickle.load(open('log33.pkl','rb')) # lr model
```

```python
std_scaler= pickle.load(open('std12.pkl','rb')) # standard scale
```

```python
st.title('Model Deployment using Logistic Regressiion')
```

```python
def user_input_parameters():
    Pregnancies = st.sidebar.number_input('Pregnancies', 0, 20)
    Glucose = st.sidebar.slider('Select your Glucose', 50, 150)
    BMI = st.sidebar.slider('Enter BMI', 10, 40)
    Age = st.sidebar.slider('Select your Age', 0, 100)
    BloodPressure = st.sidebar.number_input('BloodPressure', 80, 200)
    SkinThickness = st.sidebar.number_input('SkinThickness', 10, 30)
    Insulin = st.sidebar.number_input('Enter Insulin', 2, 25)
    DiabetesPedigreeFunction = st.sidebar.number_input(
        'Enter DiabetesPedigreeFunction', 0.0, 0.8
    )

    data = {
        'Pregnancies': Pregnancies,
        'Glucose': Glucose,
        'BloodPressure': BloodPressure,
        'SkinThickness': SkinThickness,
        'Insulin': Insulin,
        'BMI': BMI,
        'DiabetesPedigreeFunction': DiabetesPedigreeFunction,
        'Age': Age
    }

    features = pd.DataFrame(data, index=[0])

    cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness',
            'Insulin','BMI','DiabetesPedigreeFunction','Age']

    features[cols] = std_scaler.transform(features[cols])

    return features
   


df = user_input_parameters()

button = st.button('Predict')

if button:
    pred = model.predict(df)
    pred_prob = model.predict_proba(df)

    st.subheader('Prediction')
    st.write('Diabetic' if pred_prob[0][1] >= 0.5 else 'Not Diabetic')

    st.subheader('Prediction Probability')
    st.write(pred_prob)
```
