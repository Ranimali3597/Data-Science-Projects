```python
target=df['Y']
features=df.drop('Y',axis=1)
```

```python
encoder=LabelEncoder()
y=encoder.fit_transform(target)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
# Step 3: Data Splitting
# Split the data into training and test sets

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, train_size=0.8, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

3. Hyperparameter Tuning

●	Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance.
●	Adopt a structured approach like grid search or random search for hyperparameter tuning, documenting your methodology thoroughly.


```python
model = Sequential()
model.add(Dense(41, activation='relu', input_dim=60))
model.add(Dense(30, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

Compile Model

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

Fit the model

```python

#Fit the model with encoded labels and correct validation data
history = model.fit(X_train, y_train, epochs=50, batch_size=50, validation_split=0.2)
```

```python
from scikeras.wrappers import KerasClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
import tensorflow as tf
```

```python
def create_my_model1():
    model=tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(41,activation='relu',input_dim=(60)))
    model.add(tf.keras.layers.Dense(30,activation='relu'))
    model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
    model.compile(optimizer=tf.keras.optimizers.Adam(),loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
model1=KerasClassifier(build_fn=create_my_model1)
```

```python
batch_values=[10,50,60]
epochs_value=[10,50,100]
```

```python
params= dict(batch_size=batch_values,epochs=epochs_value)
params
```

```python
grid_search= GridSearchCV(estimator=model1,param_grid=params,cv=2)
```

```python
# !pip install scikit-learn==1.2.2 # Commented out to avoid potential version conflicts
```

```python
grid_search.fit(X_train,y_train)
```

```python
## summary_results
grid_search.best_score_, grid_search.best_params_
```
